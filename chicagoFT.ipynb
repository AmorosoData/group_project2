```python
#Import dependencies
import pandas as pd
import numpy as np
import pymongo
from pymongo import MongoClient
```

```python
#Define CSV path and read in CSV for Food Truck data
ft_path = "Food_Trucks.csv"
ft_df = pd.read_csv(ft_path)
ft_df.head()
```

```python
#Define CSV path and read in CSV for Census data
census_path = "IllinoisCensus_updated.csv"
census_df = pd.read_csv(census_path)
# census_df.head()
```

```python
#Rename Food Truck columns (only for columns that we will use)
ft_df.rename(columns={"DOING BUSINESS AS NAME":"business_name",
                "ZIP CODE":"zip_code",
                "APPLICATION TYPE":"app_type",
                "DATE ISSUED":"date_issued",
                "LICENSE STATUS":"license_status"},
                inplace = True)
ft_df.head()
```

```python
#Rename Food Truck columns (only for columns that we will use)
census_df.rename(columns={"zipcode":"zip_code",
                "A00100":"agi",
                "A02650":"total_income",
                "N2":"number_of_individuals",
                "NUMDEP":"number_of_dependents"},
                inplace = True)
census_df.head()
```

```python
#Create cropped dataframes for both the Food Truck dataframe and Census dataframe
crop_ft_df = ft_df[["business_name", "zip_code","app_type","date_issued", "license_status"]]
crop_census_df = census_df[["zip_code", "agi_stub", "agi", "total_income", "number_of_individuals", "number_of_dependents"]]
```

```python
#Zip codes were not the same datatypes in both dataframes(Food Truck data was a float64), so this converts to a shared datatype (int64)
crop_ft_df["zip_code"] = crop_ft_df["zip_code"].values.astype(np.int64)
```

```python
#Grouped Census dataframe by zip code, applied an aggregation function, and reset the grouped dataframe's index so that the index isn't the zipcode
group_census_df = census_df[["zip_code", "agi", "total_income", "number_of_individuals", "number_of_dependents"]]
group_census_df = group_census_df.groupby("zip_code").sum().reset_index()
group_census_df.head()
```

```python
#Used a left merge on our Census and Food Truck dataframes to merge on zip_code
merge_df = pd.merge(crop_ft_df,group_census_df, on="zip_code", how="left")
merge_df.head()
```

```python
#Pulled unique list of business names to ensure that both our unique list of businesses and length of this list is reasonable
businesses = merge_df["business_name"].unique()
n_businesses = merge_df["business_name"].nunique()
# businesses
# n_businesses
```

# Database Connection

```python
#Define connection link an instantiate client
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)

#Define the 'chicagoFT' database in Mongo
db = client.chicagoFT

#Declare the business collection
by_business = db.by_business

#Declare the zip collection
by_zip = db.by_zip
```

```python
#Instantiate for loop for populating business collection
for index, row in merge_df.iterrows():
    #Populate post dictionary with information to be inserted into business collection
    post = {
        "business_name": str(row["business_name"]),
        "Zip_Code": int(row["zip_code"]),
        "app_type": str(row["app_type"]),
        "date_issued": str(row["date_issued"]),
        "agi": str(row["agi"]),
        "total_income": str(row["total_income"]),
        "number_of_individuals": str(row["number_of_individuals"]),
        "number_of_dependents": str(row["number_of_dependents"])
    }
    #Insert data into business collection, post{} by post{}
    by_business.insert_one(post)
```

```python
#Define an interable list for our 'for loop' to loop through
unique_zip = merge_df['zip_code'].unique()

#Instantiate for loop for populating food-truck-by-zip collection
for row in unique_zip:
    #Filter dataframe to only contain zip codes equal to that which our 'for loop' is currently interating on
    df = merge_df[merge_df['zip_code'] == row]
    #Populate obj dictionary with information to be inserted into food-truck-by-zip collection. iloc[0] is used to display only 1 results, instead of a list of duplicate results
    obj = {
        'Zip-Code': str(df["zip_code"].iloc[0]),
        "Agi": str(df["agi"].iloc[0]),
        "Total_income": str(df["total_income"].iloc[0]),
        "Number_of_individuals": str(df["number_of_individuals"].iloc[0]),
        "Number_of_dependents": str(df["number_of_dependents"].iloc[0]),
        'Food_trucks' : []
    }
    #Define an interable list for our 'nested for loop' to loop through
    unique_trucks = df['business_name'].unique()
    
    #This nested 'for loop' breaks down food trucks by zip code
    for y in unique_trucks:
        #Filter dataframe to only contain business names equal to that which our 'for loop' is currently interating on
        fdf = df[df['business_name'] == y]
        #This grabs all the food truck information that we want for our current given zip code
        trucks = {
            "business_name": str(fdf["business_name"].iloc[0]),
            "app_type": str(fdf["app_type"].iloc[0]),
            "date_issued": str(fdf["date_issued"].iloc[0])
        }
        #Append food truck data onto Food_trucks list for give zip code
        obj['Food_trucks'].append(trucks)
    #Insert data into food-truck-by-zip collection, obj{} by obj{}
    by_zip.insert_one(obj)
```
